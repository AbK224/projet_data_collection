```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# Instantier l'objet chrome options
options = webdriver.ChromeOptions()
# définir l'option d'utiliser chrome en mode headless ( utiliser afin de lancer le script en background)
options.add_argument("--headless=new")
# initialiser l'instance de chrome driver en mode headless
driver = webdriver.Chrome(options=options)
```

```python
url = 'https://sn.coinafrique.com/categorie/chiens?page=1'
# contenu de la page
driver.get(url)
```

```python
# containers
containers = driver.find_elements(By.CSS_SELECTOR, '[class ="card ad__card round small hoverable  undefined "]')# div.class_name
len(containers)
```

```python
containers[0]
```

```python
#
import pandas as pd
df = pd.DataFrame()
for i in range(1,5):
    # url
    url = f'https://sn.coinafrique.com/categorie/chiens?page={i}'
    # contenu de la page
    driver.get(url)
    # containers
    containers = driver.find_elements(By.CSS_SELECTOR, '[class ="col s6 m4 l3"]')# div.class_name
    data = []
    # i = 0
    for container in containers:
        try:
            # # container
            # container = containers[0]
            # details
            nom =container.find_element(By.CSS_SELECTOR, 'p.ad__card-description').text
            # price
            price = container.find_element(By.CSS_SELECTOR, '[class = "ad__card-price"]').text
            adress = container.find_element(By.CSS_SELECTOR, '[class = "ad__card-location"]').text
            # adress
            #get image link
            image_lien = container.find_element(By.CSS_SELECTOR, '[class = "ad__card-img"]').get_attribute('src')
            dic = {
                'nom': nom,
                'price': price,
                'adress': adress,
                'image_lien': image_lien
            }
            data.append(dic)
            # i +=1
        except:
            pass

    DF = pd.DataFrame(data)
    df =pd.concat([df,DF], axis = 0).reset_index(drop= True)
```
